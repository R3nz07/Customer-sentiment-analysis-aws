```python
import boto3
import pandas as pd

# Nombre del bucket y ruta del dataset
bucket = "sentiment-analysis-ds-project"
key = "raw/amazon_reviews.csv"

# Cliente S3
s3 = boto3.client("s3")

# Descarga temporal del archivo a /tmp en la instancia
local_path = "/tmp/amazon_reviews.csv"
s3.download_file(bucket, key, local_path)

# Cargar en pandas
df = pd.read_csv(local_path)
```

```python
# Mostrar primeras filas
print("Filas totales:", len(df))
df.head()
```

```python
# Ver columnas disponibles
print(df.columns)

# Resumen de valores nulos
print(df.isnull().sum())

# Revisar ejemplos de reseñas
df.sample(5)
```
